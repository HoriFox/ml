# ЗАМЕНА ПРОПУСКОВ #

**Проверим на пропуски**

```python
print(data.isnull().sum())
```

**Наш датасет не имеет null элементов**

Если бы были пропуски, то я бы использовал заполнение числовых пропусков медианными значениями

```python
objectTypeData = data.select_dtypes(include = ['int64', 'float64']) # Только числовые значения
listOfObjectTypeData = objectTypeData.columns.tolist()

for column in listOfObjectTypeData:
    data[column] = data[column].fillna(data[column].median()) 
    # Функция median вычисляет медианные значения в столбце
    # Функция fillna заменяет NA/NAN на значение
```

Если бы были пропуски в категориальных значениях, то я бы nan и null заменил константным методом

```python
objectTypeData = data.select_dtypes(include = ['category'])
listOfObjectTypeData = objectTypeData.columns.tolist()

for column in listOfObjectTypeData:
    data[column] = data[column].fillna(data[column].describe(include=['category'])['top'])
```

```python
def plot_corr_matrix(df,size=15):
    corr = df.corr().abs()
    fig, ax = plt.subplots(figsize=(size, size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.xticks(rotation=90)
    return corr
```

```python
def statistics(test_Y, target):
    print("Средняя абсолютная ошибка:", mean_absolute_error(test_Y, target))   # Средняя абсолютная ошибка
    print("Средняя квадратичная ошибка:", mean_squared_error(test_Y, target))    # Средняя квадратичная ошибка
    print("Медианная абсолютная ошибка:", median_absolute_error(test_Y, target)) # Медианная абсолютная ошибка
```

## Разделение выборки ##

```python
from sklearn.model_selection import cross_val_score, cross_validate, train_test_split, GridSearchCV, RandomizedSearchCV, KFold, RepeatedKFold, LeaveOneOut, LeavePOut, ShuffleSplit, StratifiedKFold, learning_curve, validation_curve
from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier
from sklearn.svm import SVC, NuSVC, LinearSVC, OneClassSVM, SVR, NuSVR, LinearSVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor, AdaBoostRegressor
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error, mean_squared_log_error, median_absolute_error, precision_score, recall_score, f1_score, classification_report, confusion_matrix, roc_curve, roc_auc_score
```

```python
data_x = data.loc[:, data.columns != 'price'] # Остальные параметры
data_y = data[['price']] # Целевой признак

data_X_train, data_X_test, data_Y_train, data_Y_test = train_test_split(
    data_x, data_y, test_size=0.3, random_state=1) # Тренировочная около 70%, а тестовая для поверки 30%
data_X_train.shape, data_X_test.shape
```

* Метод k ближайших соседей
* SVM (метод опорных векторов)
* Решающие деревья
* RandomForestRegressor (Случайный лес)
* AdaBoost

## Метод ближайших соседей ##

```python
count_neighbors = 2
knr1 = KNeighborsRegressor(n_neighbors = count_neighbors)
knr1.fit(data_X_train, data_Y_train)
knr_target1 = knr1.predict(data_X_test)
print("Кол-во соседей:", count_neighbors)
statistics(data_Y_test, knr_target1)
```

```python
n_range = np.array(range(2,51,4)) # Возможные кол-во соседей
massNeighbors = [{'n_neighbors': n_range}] # Форматируем под дальнейший ввод

clf_gs = GridSearchCV(KNeighborsRegressor(), massNeighbors, cv=5, scoring='neg_mean_absolute_error') # Перекрёстная проверка
clf_gs.fit(data_X_train, data_Y_train)

print("Лучший параметр n:", clf_gs.best_params_)
```

```python
count_neighbors = clf_gs.best_params_['n_neighbors']
knr2 = KNeighborsRegressor(n_neighbors = count_neighbors)
knr2.fit(data_X_train, data_Y_train)
knr_target2 = knr2.predict(data_X_test)
print("Кол-во соседей:", count_neighbors)
statistics(data_Y_test, knr_target2)
```

## SVM (метод опорных векторов) ##

```python
svm1 = LinearSVR(max_iter=10000)
svm1.fit(data_X_train, data_Y_train)
svm_target1 = svm1.predict(data_X_test)
statistics(data_Y_test, svm_target1)
```

## Решающие деревья ## 

```python
tree1 = DecisionTreeRegressor(max_depth=2)
tree1.fit(data_X_train, data_Y_train)
tree_target1 = tree1.predict(data_X_test)
statistics(data_Y_test, tree_target1)
```

# Ансамблевые модели #

## RandomForestRegressor (Случайный лес) ##

```python
random1 = RandomForestRegressor(n_estimators=100)
random1.fit(data_X_train, data_Y_train)
random_target1 = random1.predict(data_X_test)
statistics(data_Y_test, random_target1)
```

```python
tuned_parameters = {'n_estimators': [500, 700, 1000], 'max_depth': [None, 1, 2, 3]}

forest_search = GridSearchCV(RandomForestRegressor(), param_grid=tuned_parameters, cv=5, n_jobs=-1, verbose=1)
forest_search.fit(data_X_train, data_Y_train)

print(forest_search.best_params_)
```

```python
random2 = RandomForestRegressor(random_state=1, max_depth = 3, n_estimators = 700)
random2.fit(data_X_train, data_Y_train)
random_target2 = random2.predict(data_X_test)
statistics(data_Y_test, random_target2)
```

## AdaBoost ##

```python
ab1 = AdaBoostRegressor(random_state=1, base_estimator = RandomForestRegressor(random_state=1, max_depth = 3, n_estimators = 500))
ab1.fit(data_X_train, data_Y_train)
ab_target1 = ab1.predict(data_X_test)
statistics(data_Y_test, ab_target1)
```

```python
parameters = {'n_estimators': (1, 2, 5), 'base_estimator__max_depth': (1, 2, 3)}
ab_search = GridSearchCV(ab1, parameters)
ab_search.fit(data_X_train, data_Y_train)
print(ab_search.best_params_)
```

```python
ab2 = AdaBoostRegressor(random_state=1, base_estimator = RandomForestRegressor(random_state=1, max_depth = 2, n_estimators = 1))
ab2.fit(data_X_train, data_Y_train)
ab_target2 = ab2.predict(data_X_test)
statistics(data_Y_test, ab_target2)
```
